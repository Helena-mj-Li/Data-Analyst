### Add reordered column in order_product table

For one customer, if a product is ordered for the first time, mark it as 0, otherwise mark it as 1.

Please do it in both Python and SQL. For your sql query, create a table called order_product_new which contains the reordered column.

```python
import pandas as pd
import numpy as np
```

```python
order_products = pd.read_csv("order_products.csv")
orders = pd.read_csv("orders.csv")
ordernew = order_products.merge(orders, how="left", on="order_id")
ordernew
```

```python
grouped = ordernew.groupby(by=['user_id','product_id'])
```

```python
ordernew['rank'] = grouped['order_number'].rank(ascending=True)
```

```python
ordernew
```

```python
ordernew['reordered'] = 0
ordernew.loc[ordernew['rank'] != 1] = 1
ordernew
```

```python
zero = ordernew[ordernew["reordered"] == 0]
one = ordernew[ordernew["reordered"] == 1]
print("number of reordered==0: ", zero.shape[0])
print("number of reordered==1: ", one.shape[0])
```
