Create a table called order_products_prior, you should join orders_new table with order_products_new table, filter on eval_set = prior.

```python
import numpy as np
import pandas as pd
```

```python
orders_new = pd.read_csv('orders_new.csv')
```

```python
order_products_new = pd.read_csv('order_products_new.csv')
```

```python
new = pd.merge(orders_new, order_products_new, on = 'order_id')
```

```python
new.head()
```

```python
new.info()
```

```python
order_products_prior = new[new['eval_set'] == 'prior']
```

```python
order_products_prior_csv = order_products_prior.to_csv('order_products_prior.csv', index=False)
```
