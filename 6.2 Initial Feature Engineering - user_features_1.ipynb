Create a table called user_features_1. Based on table orders_new, for each user, calculate the max order_number, the sum of days_since_prior_order and the average of days_since_prior_order.

```python
import numpy as np
import pandas as pd
```

```python
orders_new = pd.read_csv('orders_new.csv')
```

```python
orders_new.info()
```

```python
orders_new.head()
```

```python
new = orders_new.groupby(['user_id'])['order_number'].agg(['max'])
```

```python
new.head()
```

```python
new
```

```python
orders_new['user_id'].unique()
```

```python
new2 = orders_new.groupby(['user_id'])['days_since_prior_order'].agg(['sum'])
new2
```

```python
new3 = orders_new.groupby(['user_id'])['days_since_prior_order'].agg(['mean'])
new3
```

```python
user_features_1 = pd.merge(new,new2, on = 'user_id')
```

```python
user_features_1 = pd.merge(user_features_1, new3, on ='user_id')
```

```python
user_features_1
```

```python
user_features_1 = orders_new.groupby('user_id').agg(
                                                  max_order_number=('order_number',max),
                                                  sum_days_since_prior_order=('days_since_prior_order',sum),
                                                  mean_days_since_prior_order=('days_since_prior_order','mean')
)
```

```python
user_features_1
```

```python
user_features_1_csv = user_features_1.to_csv('user_features_1.csv', index=False)
```
